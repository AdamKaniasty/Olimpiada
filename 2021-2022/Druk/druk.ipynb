```python
import numpy as np
from datetime import datetime
```

```python
def test(n):
    with open("testy/in/dru{}.in".format(n),"r") as f:
        testy_in = f.read()
    with open("testy/out/dru{}.out".format(n),"r") as f:
        testy_out = f.read()
    return testy_in,testy_out
def test2(n):
    with open("testy2/in/dru{}.in".format(n),"r") as f:
        testy_in = f.read()
    with open("testy2/out/dru{}.out".format(n),"r") as f:
        testy_out = f.read()
    return testy_in,testy_out
    
def data(n,t):
    if t == 1:
        t_in,t_out = test(n)
        t_in = t_in.split("\n")[:-1]
        n,m = t_in[0].split(" ")
        t_out = t_out.split("\n")[:-1]
        t_out[1] = t_out[1].split(" ")[:-1]
        t_in = t_in[1:-1]
        if len(t_in) == int(n) and len(t_in[0]) == int(m):
            return int(n),int(m),t_in,t_out
        else:
            return False
    if t == 2:
        t_in,t_out = test2(n)
        t_in = t_in.split("\n")[:-2]
        n,m = t_in[0].split(" ")
        t_out = t_out.split("\n")[:-1]
        t_out[1] = t_out[1].split(" ")[:-1]
        t_in = t_in[1:]
        if len(t_in) == int(n) and len(t_in[0]) == int(m):
            return int(n),int(m),t_in,t_out
        else:
            return False
```

```python
def letter_check(l,c):
    lc = {}
    for i in c:
        try:
            lc[i] += 1
        except:
            lc.update({i:1})
    s = 0
    for i in l.keys():
        if s == 0:
            s = l[i]/lc[i]
        else:
            try:
                if l[i]/lc[i] != s:
                    return False
            except KeyError:
                return False
    return True
```

```python
def count(t,T,c):
    w = 0 
    l = len(c)
    for i in range(0,len(t),1):
        j = 0
        while j <= len(t[i]):
            if t[i][j:j+l] == c:
                w += 1
                j += l-1
            j = j+1
    for i in range(0,len(T),1):
        j = 0
        while j <= len(T[i]):
            if T[i][j:j+l] == c:
                w += 1
                j += l-1
            j = j+1
    return w

def check(t,T,p,i,l):
    target = int(p/i)
    c = t[0][0:i]
    if letter_check(l,c):
        cnt = count(t,T,c)
        if target <= cnt:
            return True
    return False
        
```

```python
def main(test_nr,test_type):
    n,m,t,out = data(test_nr,test_type)
    # print(n,m,t)
    start = datetime.now()
    w = []
    temp = [list(x) for x in t]
    T = np.asarray(temp).T.tolist()
    T = ["".join(x) for x in T]
    l = {}
    for i in t:
        for j in i:
            try:
                l[j] += 1
            except:
                l.update({j:1})
    if len(l.keys()) == 1:
        s = set()
        ma = n if n > m else m
        for i in range (1,int(ma/2)+1):
            if n%i == 0:
                s.add(i)
            elif m%i == 0:
                s.add(i)
        s.add(n)
        s.add(m)
        s = list(s)
        s.sort()
        end = datetime.now()
        return (len(s),s),out,end-start,l
    else:
        p = m*n
        ma = n if n>m else m
        for i in range(2,ma+1):
            if p%i == 0:
                if i <= n and i <= m:
                    if check(t,T,p,i,l) == True:
                        w.append(i)
                    elif check(T,t,p,i,l) == True:
                        w.append(i)
                if i <= n and i > m:
                    if check(T,t,p,i,l) == True:
                        w.append(i)
                if i > n and i <= m:
                    if check(t,T,p,i,l) == True:
                        w.append(i)
    end = datetime.now()
    return (len(w),w),out,end-start,l
```
