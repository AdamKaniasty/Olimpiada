```python
from math import factorial 
import numpy as np
from itertools import permutations
```

```python
def test(n):
    with open ("testy/in/imp{}.in".format(n), "r") as f:
        test_in = f.read()
    with open ("testy/out/imp{}.out".format(n), "r") as f:
        test_out = f.read()
    test_in = test_in.split("\n")
    test_in[1] = test_in[1].split(" ")[:-1]
    return test_in,test_out
```

```python
def modulo(n):
    return n%10000007
def certainty(n,t):
    w = np.zeros(n)
    z = np.zeros(n)
    z[0] = 1
    z[-1] = 1
    w[1]=t[0]
    w[-2]=t[-1]
    for i in range(0,n):
        if w[i] == 0:
            if i != 0 and i != n-1:
                if t[i-1] == t[i+1]:
                    w[i] = t[i-1]
                    z[i-1] = 1
                    z[i+1] = 1
                elif w[i+2] != 0 and w[i+2] != t[i+1]:
                    w[i] = t[i+1]
                    z[i+1] = 1
                elif w[i-2] != 0 and w[i-2] != t[i-1]:
                    w[i] = t[i-1]
                    z[i-1] =1
            if i == n-1:
                if w[i-2] != 0 and w[i-2] != t[i-1]:
                    w[i] = t[i-1]
                    z[i-1] =1
            if i == 0:
                if w[i+2] != 0 and w[i+2] != t[i+1]:
                    w[i] = t[i+1]
                    z[i+1] =1
        else:
            try:
                if t[i-1] == w[i]:
                    z[i-1] = 1
            except IndexError:
                pass
            try:
                if t[i+1] == w[i]:
                    z[i+1] = 1
            except IndexError:
                pass
    for i in range(n-1,-1,-1):
        if w[i] == 0:
            if i != 0 and i != n-1:
                if t[i-1] == t[i+1]:
                    w[i] = t[i-1]
                    z[i-1] = 1
                    z[i+1] = 1
                elif w[i+2] != 0 and w[i+2] != t[i+1]:
                    w[i] = t[i+1]
                    z[i+1] = 1
                elif w[i-2] != 0 and w[i-2] != t[i-1]:
                    w[i] = t[i-1]
                    z[i-1] =1
            if i == n-1:
                if w[i-2] != 0 and w[i-2] != t[i-1]:
                    w[i] = t[i-1]
                    z[i-1] =1
            if i == 0:
                try:
                    if w[i+2] != 0 and w[i+2] != t[i+1]:
                        w[i] = t[i+1]
                        z[i+1] =1
                except:
                    pass
    return w,z

def daszki(n,z):
    w = []
    m = np.zeros(n)
    total = 0
    for i in range(n):
        cnt = 0
        if z[i] == 0 and m[i] != 1:
            # print("ROZPATRUJE",i)
            m[i] = 1
            cnt = 1
            total += 1
            try:
                if z[i+2] == 0:
                    # print("WIECEJ NIZ 1")
                    for j in range(i+2,n,2):
                        if z[j] == 0:
                            m[j] = 1
                            cnt += 1
                            total += 1
                        else:
                            break
            except IndexError:
                pass
            # print("APPENDING")
            w.append(cnt)
 
    return w,total
                    
def main(n,t):
    w,z = certainty(n,t)
    d,total_zero = daszki(n,z)
    # print("WP≈ÅYW",z)
    # print("DASZKI",d)
    q =np.count_nonzero(w==0)
    scale = 1
    increment = 1
    last_factorial = factorial(q-total_zero)
    for i in d:
        for j in range(1,i+1):
            scale += increment
            # print(i,j,scale)
        increment = scale
        
    output = modulo(scale) * modulo(last_factorial)
    # print(scale,last_factorial)
    return modulo(output)
```

```python
res = test(21)
n = int(res[0][0])
t = [int(x) for x in res[0][1]]
test_out = res[1]
output = main(n,t)
print(output)
print(test_out)
```

```python
for i in range(1,57):
    res = test(i)
    n = int(res[0][0])
    t = [int(x) for x in res[0][1]]
    test_out = res[1]
    output = main(n,t)
    if(int(output) != int(test_out)):
        print(i)
        print(output,test_out)
```
