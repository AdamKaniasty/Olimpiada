```python
def test(n):
    with open ("testy/in/imp{}.in".format(n), "r") as f:
        test_in = f.read()
    with open ("testy/out/imp{}.out".format(n), "r") as f:
        test_out = f.read()
    test_in = test_in.split("\n")
    test_in[1] = test_in[1].split(" ")[:-1]
    return test_in,test_out
def test2(n):
    with open ("testy2/in/imp{}.in".format(n), "r") as f:
        test_in = f.read()
    with open ("testy2/out/imp{}.out".format(n), "r") as f:
        test_out = f.read()
    test_in = test_in.split("\n")
    test_in[1] = test_in[1].split(" ")
    return test_in,test_out

def test3(n):
    with open ("testy3/in/{}.in".format(n), "r") as f:
        test_in = f.read()
    with open ("testy3/out/{}.out".format(n), "r") as f:
        test_out = f.read()
    test_in = test_in.split("\n")
    test_in[1] = test_in[1].split(" ")[:-1]
    return test_in,test_out
```

```python
def modulo(n):
    return n%1000000007

def certainties(n,t):
    if n == 3 and t[0] != t[2]:
        return False
    
    w = np.zeros(n)
    z = np.zeros(n)
    occ = np.zeros(n+1)
    w[1] = t[0]
    w[-2] = t[-1]
    z[0] = 1
    z[-1] = 1
    
    for i in range(1,n-1):
        if i == 1:
            occ[t[0]] += 1
            occ[t[n-1]] += 1
        occ[t[i]] += 1
        a = t[i-1]
        b = t[i+1]
        c = w[i]
        if a == b:
            if w[i] != a and w[i] != 0:
                return False
            w[i] = a
            z[i+1] = 1
            z[i-1] = 1
        elif a == c:
            try:
                if w[i+2] != 0 and w[i+2] != b:
                    return False
                w[i+2] = b
                z[i+1] = 1
            except IndexError:
                pass
        elif b == c:
            try:
                if w[i-2] != 0 and w[i-2] != a:
                    return False
                w[i-2] = a
                z[i-1] = 1
            except IndexError:
                pass   
    for i in range(n-2,0,-1):
        a = t[i-1]
        b = t[i+1]
        c = w[i]
        if a == b:
            w[i] = a
            z[i+1] = 1
            z[i-1] = 1
        elif a == c:
            try:
                w[i+2] = b
                z[i+1] = 1
            except IndexError:
                pass
        elif b == c:
            try:
                w[i-2] = a
                z[i-1] = 1
            except IndexError:
                pass  
    if max(occ) > 2:
        return False
    return w,z
    
def daszki(n,z):
    w = []
    m = np.zeros(n)
    total = 0
    for i in range(n):
        cnt = 0
        if z[i] == 0 and m[i] != 1:
            m[i] = 1
            cnt = 1
            total += 1
            try:
                for j in range(i+2,n,2):
                    if z[j] == 0:
                        m[j] = 1
                        cnt += 1
                        total += 1
                    else:
                        break
            except IndexError:
                pass
            w.append(cnt)
 
    return w,total

def main(n,t):
    c = certainties(n,t)
    if c == False:
        return 0
    else:
        w,z = c
    # print("T",t)
    # print("W",w)
    # print("Z",z)
    d,total = daszki(n,z)
    # print(d,total)

    q = np.count_nonzero(w==0)
    scale = 1
    increment = 1
    last_factorial = factorial(q-total)
    for i in d:
        for j in range(0,i):
            scale += increment
        increment = scale
    output = modulo(modulo(scale) * modulo(last_factorial))
    return modulo(output)
```

```python
time = []
for num in range(2377,50000):
    inp,out = test3(num)
    n,t = inp
    n = int(n)
    t = [int(x) for x in t]
    t = np.array(t)
    start = datetime.now()
    program = main(n,t)
    end = datetime.now()
    time.append(end-start)
    print(num,end-start)
    if program != int(out):
        print("ERROR",num,program,int(out))
        print()
```
