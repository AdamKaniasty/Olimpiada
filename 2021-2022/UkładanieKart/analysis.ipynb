```python
from itertools import permutations
import numpy as np
from math import factorial
```

```python
def sorted(cards):
    return all(cards[i]<cards[i+1] for i in range(len(cards)-1))
def time(cards,n):
    total = 0
    t = np.zeros(n+1)
    t2 = np.zeros(n+1)
    s = sorted(cards)
    if s == True:
        return 0
    while 1:
        a = cards[0]
        if a != 1:
            b = cards.index(a-1)
            total += b
            t[a-1] += b
            t2[a-1] +=1
            del(cards[b])
            cards.insert(0,a-1)
        else:
            b = cards.index(n)
            total += b
            t[n] += b
            t2[n] += 1
            del(cards[b])
            cards.insert(0,n)
        if sorted(cards):
            break    
    return t,t2

def permutation(n):
    t = np.arange(1,n+1,1)
    weight = np.zeros(n+1)
    t2 = np.zeros(n+1)
    perms = list(permutations(t.tolist()))
    total = 0
    ma = 0
    for i in perms:
        output = time(list(i),n)
        try:
            for i in range(len(output[0])):
                weight[i] += output[0][i]
                t2[i] += output[1][i]
        except TypeError:
            pass
    return (total,ma,weight,t2)
```

```python
scores = []
r=8
for i in range(1,r+1):
    out = permutation(i)
    weights =  out[2][1:]
    occ = out[3][1:]
    print(i,weights,occ)
    try:
        print("R:",int(weights[0]-weights[1]))
        print("Średnia:",sum(weights)/i)
        print("Silnia:",factorial(i))
        print("R-wystąpień",int(occ[0]-occ[1]))
        print("Suma-wystąpień",int(sum(occ)))
        print("Suma",int(sum(weights)))
    except:
        pass
    print()
    scores.append([i, out[0],out[1],factorial(i)])
```
