```python
from itertools import permutations
import numpy as np
from math import factorial
```

```python
def sorted(cards):
    return all(cards[i]<cards[i+1] for i in range(len(cards)-1))
def time(cards,n):
    total = 0
    t = 0
    s = sorted(cards)
    if s == True:
        return 0
    while 1:
        a = cards[0]
        if a != 1:
            b = cards.index(a-1)
            total += b
            del(cards[b])
            cards.insert(0,a-1)
        else:
            b = cards.index(n)
            total += b
            t = b
            del(cards[b])
            cards.insert(0,n)
        if sorted(cards):
            break    
    return (t)

def permutation(n):
    t = np.arange(1,n+1,1)
    perms = list(permutations(t.tolist()))
    total = 0
    t2 = np.zeros(n+1)
    for i in perms:
        output = time(list(i),n)
        try:
            total += output
            if output != 0:
                print(i[0],output)
                t2[i[0]] += 1

        except:
            pass
    return (total,t2[1:])
```
